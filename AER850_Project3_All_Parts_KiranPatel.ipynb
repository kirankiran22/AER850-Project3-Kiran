```python
# AER850 - Project 3
# By: Kiran Patel
# Dec 12th 2024

# Importing Libraries
#!pip install ultralytics
import matplotlib.pyplot as plt
import cv2
import numpy as np

# Step 1: Object Masking
#------------------------------------------------------------------------------

# Mount Google Drive to access files
from google.colab import drive
drive.mount('/content/drive')


# Path to the image file stored in Google Drive
image_path = '/content/drive/MyDrive/ColabNotebooks/motherboard_image.JPEG'

# Load the image
original_image = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Rotate the image 90 degrees clockwise for proper orientation
rotated_image = cv2.rotate(original_image, cv2.ROTATE_90_CLOCKWISE)

# Apply a uniform blur to reduce noise and smooth the image
blurred_image = cv2.blur(rotated_image, (15, 15))

# Convert the blurred image to greyscale for making the thresholding process
# Common practise in image processing to greyscale first
greyscale_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)


# Apply adaptive thresholding for better segmentation
# Maximum intensity value (255 for white)
thresholded_image = cv2.adaptiveThreshold(
    greyscale_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 55, 10
)

# Detect edges using the Canny edge detector
edges_image = cv2.Canny(thresholded_image, 35, 175)

# Dilate the edges to make them thicker and fill gaps
dilated_edges = cv2.dilate(edges_image, None, iterations = 12)

# Find contours in the dilated edge image
contours, _ = cv2.findContours(dilated_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a blank mask with the same dimensions as the original image
mask = np.zeros_like(rotated_image)

# Find the largest contour based on area
largest_contour = max(contours, key=cv2.contourArea, default = None)

# Draw the largest contour onto the mask
cv2.drawContours(mask, [largest_contour], -1, (255, 255, 255), thickness=cv2.FILLED)

# Apply the mask to the original rotated image to extract the PCB
masked_image = cv2.bitwise_and(rotated_image, mask)

# Results
#------------------------------------------------------------------------------

# Display the edge-detected image
plt.figure()
plt.imshow(edges_image, cmap="gray")
plt.title("Edge Detection Results", color = 'blue')

# Display the mask
plt.figure()
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
plt.title("Masking Results", color = 'green')

# Display the final masked PCB image
plt.figure()
plt.imshow(cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB))
plt.title("Masked Image - Final Result", color = 'red')

# Show all the plots
plt.show()

#Save this result
output_path = "/content/drive/My Drive/ColabNotebooks/Step1-Outputs/masked_motherboard.jpg"
cv2.imwrite(output_path, masked_image)
print(f"Masked image saved to: {output_path}")

# Print the dimensions of the original and masked images
print("Original Image Shape:", rotated_image.shape)
print("Masked Image Shape:", masked_image.shape)
#------------------------------------------------------------------------------

#Step 2 - YOLO V8 Training
#------------------------------------------------------------------------------
# Install YOLOv8
from ultralytics import YOLO

#Defining Dataset Paths
dataset_path = '/content/drive/MyDrive/ColabNotebooks/data/data.yaml'
output_path = '/content/drive/MyDrive/ColabNotebooks/Step2-Outputs/'

# Load YOLOv8 Nano model
model_1 = YOLO('yolov8n.pt')

results = model_1.train(
    data= dataset_path,              # Path to dataset configuration file
    epochs = 200,                    # Number of training epochs
    batch = 2,                       # Batch size
    imgsz = 1280,                    # Image size
    name = 'Project3_YOLO_Training', # Name of the run
    plots = True,                    # Save plots of training metrics
    project = output_path,           # Output directory
)

# Step 3: YOLOv8 Evaluation
# ------------------------------------------------------------------------------
# Paths to evaluation images
arduinomega_path = "/content/drive/MyDrive/ColabNotebooks/data/evaluation/ardmega.jpg"
arduinoUno_path = "/content/drive/MyDrive/ColabNotebooks/data/evaluation/arduno.jpg"
rasberrypi_path = "/content/drive/MyDrive/ColabNotebooks/data/evaluation/rasppi.jpg"

# Prediction for Ardmega
results_arduinomega = model_1.predict(
    source = arduinomega_path,  # Path to Ardmega image
    imgsz = 2048,               # Image size
    conf = 0.35,                # Confidence threshold
    save = True,                # Save the prediction
)

# Prediction for Ardu
results_arduinoUno = model_1.predict(
    source=arduinoUno_path,  # Path to Ardu image
    imgsz = 2048,            # Image size
    conf=0.35,               # Confidence threshold
    save=True,               # Save the prediction
)

# Prediction for Raspi
results_rasberrypi = model_1.predict(
    source=rasberrypi_path,  # Path to Raspi image
    imgsz = 2048,            # Image size
    conf=0.35,               # Confidence threshold
    save=True,               # Save the prediction
)
#------------------------------------------------------------------------------
```
